```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv("Datasets/ColourXY.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
X = data
```

```python
plt.scatter(X['x'],X['y'],c = data['color'].astype('category'))
```

```python
X = data.drop(['color'],axis=1)
```

```python
X.head()
```

```python
help(KMeans)
```

```python
wcc = []
for i in range(1,10):
    kmeans = KMeans(n_clusters=i,init="k-means++",verbose=True)
    kmeans.fit_predict(X)
    wcc.append(kmeans.inertia_)
```

```python
plt.plot(range(1,10),wcc)
plt.show()
```

```python
kmeans = KMeans(n_clusters=3,init="k-means++",verbose=True)
kmeans.fit_predict(X)
```

```python
print(kmeans.cluster_centers_)
```

```python
print(kmeans.labels_)
```

```python
plt.scatter(X['x'],X['y'],c = data['color'].astype('category'))
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],marker='s',c=['g','r','y'],s=200)
plt.legend()
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("KMeans Scattering Plot")
```
